```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
books=pd.read_csv('book.csv',encoding='Latin1')
books
```

```python
books1=books.drop("Unnamed: 0", axis=1)
books1
```

```python
books1.sort_values(['User.ID'], inplace=True)
```

```python
len(books1['User.ID'].unique())
```

```python
books1["User.ID"].value_counts()
```

```python
len(books1["User.ID"].value_counts())
```

```python
len(books1['Book.Title'].unique())
```

```python
books2=books1.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
books2
```

```python
books2.index=books1['User.ID'].unique()
books2
```

```python
books2.fillna(0,inplace=True)
books2
```

```python
user_similarity=1-pairwise_distances(books2.values,metric='cosine')
user_similarity
```

```python
pd.DataFrame(user_similarity)
```

```python
user_similarity
```

```python
user_similarity1=pd.DataFrame(user_similarity)
user_similarity1
```

```python
user_similarity1.index=books1['User.ID'].unique()
user_similarity1.columns=books1['User.ID'].unique()
user_similarity1
```

```python
np.fill_diagonal(user_similarity, 0)
user_similarity1
```

```python
user_similarity
```

```python
user_similarity1.idxmax(axis=1)[0:5]
```

```python
for i in user_similarity1.index:
    if user_similarity1.loc[i].max()!=0:
        print(i, "and", user_similarity1.idxmax(axis=1)[i], ":", user_similarity1.loc[i].max())
```
